```python
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
data['keywords'] = data['keywords'].map(lambda x: ast.literal_eval(x))
```

```python
def genresList(x):
    gen = []
    st = " "
    for i in x:
        gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['genres'] = data['genres'].map(lambda x: genresList(x))
data['genres']
```

```python
s = data.apply(lambda x: pd.Series(x['keywords']),axis=1).stack().reset_index(level=1, drop=True)
s.name = 'keyword'
s = s.value_counts()
s[:5]
```

```python
s = s[s > 1]
stemmer = SnowballStemmer('english')
```

```python
def keywordsFiltered(x):
    words = []
    for i in x:
        if i in s:
            words.append(i)
    return words
```

```python
data['keywords'] = data['keywords'].apply(keywordsFiltered)
data['keywords'] = data['keywords'].apply(lambda x: [stemmer.stem(i) for i in x])
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
          return str.lower(i['name'].replace(" ", ""))
    return np.nan
```

```python
data['director'] = data['crew'].apply(get_director)
```
