```python
links = pd.read_csv('/content/drive/My Drive/Movie/links_small.csv')
links = links[links['tmdbId'].notnull()]['tmdbId'].astype('int')
def convert_int(x):
    try:
        return int(x)
    except:
        return np.nan

moviesMeta['id'] = moviesMeta ['id'].apply(convert_int)
moviesMeta[moviesMeta ['id'].isnull()]
```

```python
moviesMeta = moviesMeta.drop([19730, 29503, 35587])
moviesMeta['id'] = moviesMeta['id'].astype('int')
data = moviesMeta[moviesMeta['id'].isin(links)]
data.shape
```

```python
data = data.loc[:,['id','overview','tagline','title']]
data
```

```python
data[data['tagline'].isnull()]
```

```python
data['tagline'] = data['tagline'].fillna('')
```

```python
data['comb'] = data['overview'] + data['tagline']
data['comb'] = data['comb'].fillna('')
```
